## 1. Wave-Particle Duality

The question we intend to answer is: "Is the light a wave or a particle?"

### The light shows properties of both:
#### As particle:

- a) Reflection: Light bounces off surfaces at equal angles; explained by wave fronts (Huygens' principle).

 Law of Reflection: $ \theta_i = \theta_r $ (angle of incidence equals angle of reflection).

- b) Refraction: Light bends when changing mediums due to speed change; also via Huygens' principle.

 Snell's Law: $ n_1 \sin\theta_1 = n_2 \sin\theta_2 $ (where $ n $ is refractive index).


#### As wave:
- a) Photoelectric Effect: Light ejects electrons from a metal surface only if its frequency exceeds a threshold, regardless of intensity. This supports photons transferring energy in quanta.
Formula: Maximum kinetic energy of ejected electrons: $ KE_{\max} = hf - \phi $ (where $ \phi $ is the work function of the metal).


- b) Diffraction: Light bends around edges or through small openings, creating patterns.

Formula (single-slit minimum): $ a \sin\theta = m\lambda $ (where $ a $ is slit width, $ m = 1, 2, \dots $).

- c) Interference: Light waves overlap to create bright (constructive) and dark (destructive) patterns, e.g., in Young's double-slit experiment.

"Interference is a phenomenon in which two waves superpose to form a resultant wave.
The resultant wave can have greater, lower, or the same amplitude, depending on the phases of the interfering waves."

Formula: Path difference for constructive interference: $ \delta = m\lambda $ (where $ m = 0, 1, 2, \dots $; for destructive: $ \delta = (m + 1/2)\lambda $).

### Conclusion:

**Electrons (and all microscopic particles) display properties of both waves and particles.**





```python
import matplotlib.pyplot as plt
import numpy as np

# Law of Reflection: $θ_incident$ = $θ_reflected$
# Simple visualization: Incident ray hits a surface (horizontal line) and reflects.

# Parameters
theta_i = 45  # Incident angle in degrees
theta_r = theta_i  # Reflected angle (formula: θ_r = θ_i)

# Convert to radians for trig
theta_i_rad = np.deg2rad(theta_i)
theta_r_rad = np.deg2rad(theta_r)

# Surface position (horizontal line at y=0)
x_surface = np.linspace(-1, 1, 100)
y_surface = np.zeros(100)

# Incident ray: from ( -1, 1 ) towards origin, based on angle
# Point of incidence at (0,0)
incident_start = (-np.cos(theta_i_rad), np.sin(theta_i_rad))  # Normalize to length 1
incident_end = (0, 0)

# Reflected ray: from origin at reflected angle
reflected_start = (0, 0)
reflected_end = (np.cos(theta_r_rad), np.sin(theta_r_rad))  # Note: reflection flips x-direction

# Normal line (perpendicular to the surface)
normal_x = [0, 0]
normal_y = [0, 0.5]

# Plot
plt.figure(figsize=(6, 4))
plt.plot(x_surface, y_surface, 'k-', label='Surface') 
plt.plot([incident_start[0], incident_end[0]], [incident_start[1], incident_end[1]], 'b-', label='Incident Ray')
plt.plot([reflected_start[0], reflected_end[0]], [reflected_start[1], reflected_end[1]], 'r-', label='Reflected Ray')
plt.plot(normal_x, normal_y, 'g--', label='Normal')

# Labels
plt.text(-0.5, 0.5, f'$θ_i$ = {theta_i}°', color='blue')
plt.text(0.2, 0.5, f'$θ_r$ = {theta_r}°', color='red')

plt.xlim(-1.5, 1.5)
plt.ylim(-0.5, 1.5)
plt.axis('equal')
plt.legend()
plt.title('Law of Reflection: $θ_i$ = $θ_r$')
plt.show()
```

## 2.Quantization

It started in 1900 - **Max Plank**.
He suggested that matter — for example, electrons — doesn’t give off radiation of frequency f in a continuous range of energies. Instead, it can only transfer energy in whole-number chunks of a specific size, what we now call a quantum of energy.


### Energy Quantization Principle

- Photon quantized energy:  
  $$ E = h f $$  or equivalently  $$ E = \hbar \omega  \tag{1} $$

- Angular frequency relation:  
  $$ \omega = 2\pi f \tag{2} $$

- Reduced Planck's constant:  
  $$ \hbar = \frac{h}{2\pi} $$

- Planck’s constant:  
  $$ h = 6.626 \times 10^{-34} \, \mathrm{J \cdot s} \tag{3} $$

- Reduced Planck’s constant:  
  $$ \hbar = 1.055 \times 10^{-34} \, \mathrm{J \cdot s} $$
